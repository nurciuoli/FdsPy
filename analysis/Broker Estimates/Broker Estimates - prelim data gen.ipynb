```python
import sys
import os
import numpy as np
import pandas as pd
sys.path.append(os.path.dirname(os.path.dirname(os.path.dirname(os.path.abspath('"Top 10 Funds by AUM Positioning - Data Gen.ipynb"')))))
sys.path.append(os.path.dirname(os.path.dirname(os.path.dirname(os.path.abspath('"Top 10 Funds by AUM Positioning - Data Gen.ipynb"')))))
import apis.qe.myqengine as qe
```

```python
time_series = qe.TimeSeries(start_date = '0M',end_date = '0M', frequency='M')

univ = qe.ScreenUniverse(universe_expr = 'FG_CONSTITUENTS(DJIAK,0,CLOSE)=1')
req = qe.calculate(universe = univ, dates = time_series,formulas = ["FE_BROKER_ESTIMATE(SNAP,EPS,BKR_NAME,ANN_ROLL,+1,#DATE,,,,'WIN=31')", #not using _ROLL because I want to focus on FY 24
                                            "FE_BROKER_ESTIMATE(SNAP,EPS,AN_NAME,ANN_ROLL,+1,#DATE,,'WIN=31')",
                                           "FE_BROKER_ESTIMATE(SNAP,EPS,EST_VALUE,ANN_ROLL,+1,#DATE,,'WIN=31')",
                                           "FE_BROKER_ESTIMATE(SNAP,EPS,PREV_VALUE,ANN_ROLL,+1,#DATE,,'WIN=31')",
                                        "FE_BROKER_ESTIMATE_DATE(SNAP,EPS,ESTDATEN,ANN_ROLL,+1,'MM/DD/YYYY',#DATE,,'WIN=31')"],
                                        source = 'FqlExpression',
                                        is_array_return_type=True)
```

```python

df = req.data
df.head()
```

```python
df= df.set_index(['DATE',"UNIVERSE"])
```

```python
df= df.rename(columns = {"FE_BROKER_ESTIMATE(SNAP,EPS,BKR_NAME,ANN_ROLL,+1,#DATE,,,,'WIN=31')":"Broker Name",
                     "FE_BROKER_ESTIMATE(SNAP,EPS,AN_NAME,ANN_ROLL,+1,#DATE,,'WIN=31')":'Analyst Name',
                    "FE_BROKER_ESTIMATE(SNAP,EPS,EST_VALUE,ANN_ROLL,+1,#DATE,,'WIN=31')":"EST_EPS_FY1",
                    "FE_BROKER_ESTIMATE(SNAP,EPS,PREV_VALUE,ANN_ROLL,+1,#DATE,,'WIN=31')":"PREV_EST_EPS_FY1",
                    "FE_BROKER_ESTIMATE_DATE(SNAP,EPS,ESTDATEN,ANN_ROLL,+1,'MM/DD/YYYY',#DATE,,'WIN=31')":'Research Date'})
```
