## Import Lib and Set Inputs

```python
import myfpe as fpe
```

## Get Time Series

```python
start = '-11M'
stop = '-10M'
freq= 'M'
cal ='NAY'
req= fpe.get_time_series(start,stop,freq,cal)
req.data
```

## Get Universe Constituents

```python
universe_formula = 'FG_CONSTITUENTS(OEX,0,CLOSE)=1'
fpe.get_univ_constituents(universe_formula,
                            start,
                            stop,
                            freq,
                            cal).data
```

## My Screen and Universe Calc

```python
formulas = ['P_PRICE(0,USD)','P_PRICE_RETURNS(1,-1/0/0,0,USD)']
columns = ['price','tret']

fpe.screen_scr_univ(universe_formula=universe_formula,
                    start=start,
                    stop=stop,
                    freq=freq,
                    cal=cal,
                    formulas=formulas,
                    columns=columns).data
```

## FQL and Universe Calc

```python
fql_formulas = ['P_PRICE(#DATE,,,USD)']
fql_columns = ['price']

fpe.fql_scr_univ(universe_formula=universe_formula,
                start=start,
                stop=stop,
                freq=freq,
                cal=cal,
                formulas=fql_formulas,
                columns=fql_columns).data
```

## Screen/FQL with Known IDs

### Screening

```python
ids = ['AAPL','MSFT','AMZN']
data = fpe.screen_ids(ids = ids,
                    start=start,
                    stop=stop,
                    freq=freq,
                    cal=cal,
                    formulas=formulas,
                    columns=columns).data
data
```

### FQL 

```python
fpe.fql_ids(ids = ids,
            start=start,
            stop=stop,
            freq=freq,
            cal=cal,
            formulas=fql_formulas,
            columns=fql_columns).data
```

## Archive & Retrieve

### Upload file to FPE

```python
path= 'testing_file_archive.csv'
fpe.post_file(path,data)
```

### Retrieve file archived in FPE

```python
fpe.retrieve_file(path).data
```
