```python
from FdsPy.content.utils import id_lookup,search_people,search_symbols
```

```python
id_lookup('Amazon',search_type=['equities'],result_limit=5)
```

```python
people_match_request ={
    "input": [{"personName":"Andy Jassy",'entity' :'Amazon'},
               {"personName":'Tim Cook','entity':'Apple'}],
    }

search_people(people_data=people_match_request)
```

```python
entity_match_request ={
    "input": [{"name":"Apple Inc.","clientId": "abc-123"},
    {"name":"Microsoft Corporation", "clientId": "abc-123"},
    {"name":"Amazon.com, Inc.", "clientId": "abc-123"},
    {"name":"Tesla Inc", "clientId": "abc-123"},
    {"name":"Alphabet Inc. Class A", "clientId": "abc-123"},],
    "includeEntityType": ["PUB"],
    "excludeEntityType": ["EXT"],
    "includeEntitySubType": ["PR"],
    "excludeEntitySubType": ["AR"],
    "includeParent": False
    }

search_symbols(entity_data=entity_match_request)
```
